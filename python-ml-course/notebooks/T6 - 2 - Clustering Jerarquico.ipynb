# Clustering jerárquico y dendrogramas
Antes de empezar, pongamos un poco de notación para hablar todos el mismo idioma

* X dataset (array de n x m) de puntos a clusterizar
* n número de datos
* m número de rasgos 
* Z array de enlace del cluster con la información de las uniones
* k número de clusters

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram,linkage
import numpy as np
```

```python
np.random.seed(4711)
# genero aleatoriamente mis datos
a = np.random.multivariate_normal([10,0],[[3,1],[1,4]], size = [100,])
b = np.random.multivariate_normal([0,20], [[3,1],[1,4]], size = [50,])
X = np.concatenate((a,b))
print(X.shape)
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
Z = linkage(X, "ward")
```

```python
# Z es la matriz de enlace
# coeficiente de cophenet
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist
```

```python
# le paso a la funcion la matriz de enlace y las distancias originales (pdist)
c, coph_dist = cophenet(Z, pdist(X))
# c es el % de conservación de las distancias entre la matriz de enlace y las dist originales
c
```

### Dendrograma tuneado
para mostrar en el endograma la distancia de unión de cada cluster

```python
def dendrogram_tune(*args, **kwargs):
    
    max_d=kwargs.pop("max_d", None)
    if max_d and 'color_threshold' not in kwargs:
        kwargs['color_threshold'] = max_d
    annotate_above = kwargs.pop('annotate_above', 0)
    
    ddata = dendrogram(*args,**kwargs)
        
    if not kwargs.get('no_plot', False):
        plt.title("Clustering jerárquico con Dendrograma truncado")
        plt.xlabel("Índice del Dataset (o tamaño del cluster)")
        plt.ylabel("Distancia")
        for i, d, c in zip(ddata['icoord'], ddata['dcoord'], ddata['color_list']):
            x = 0.5 * sum(i[1:3])
            y = d[1]
            if y>annotate_above:
                plt.plot(x,y,'o',c=c)
                plt.annotate('%.3g'%y, (x,y), xytext=(0,-5),
                            textcoords="offset points", va="top", ha="center")
                
    if max_d:
        plt.axhline(y=max_d, c='k')
        
    return ddata
```

```python
dendrogram_tune(Z,truncate_mode='lastp',p=12, leaf_rotation=90., leaf_font_size=12.,
                show_contracted=True,annotate_above=10, max_d=13)
plt.show()

# Con max_d especifico mi corte
```

### Corte automático del dendrograma
Método de la inconsistencia

inconsistency_i = (h_i-avg(h_j))/std(h_j)

trata de ajustar cada una de las inconsistencias dividido por la desviación estandar de cada una de las uniones

```python
from scipy.cluster.hierarchy import inconsistent
```

```python
depth = 3 # los depth niveles anteriores especificados del deneagrama
incons = inconsistent(Z, depth) # buscamos las inconsistencias con profundidad 3
incons[-10:]

# promedio, desviación estandar, count y factor de inconsistencia

# no es muy util porque depende de los niveles anteriores
```

### Método del codo


```python
last = Z[-10:,2] # me quedo con las ultimas 10
last_rev = last[::-1] # los revierto, por lo que al principio va a estar el ultimo cluster
print(last_rev)
idx = np.arange(1, len(last)+1) # indices para rep grafica
plt.plot(idx, last_rev)

# calculamos las aceleraciones
acc = np.diff(last,2) # diferencias
acc_rev = acc[::-1] # 1 acc menos que los datos originales
plt.plot(idx[:-2]+1, acc_rev)
plt.show()
k = acc_rev.argmax() +2 # buscamos el max de las aceleraciones + 2 clusters (???)
print("El número óptimo de cluster es %s"%str(k))


# intenta buscar el step del clustering donde la aceleración del crecimiento de las distancias es mayor

# el codo no funciona para decir que se tiene 1 cluster o N, porque no permite medir diferencias de acc
```

## Recuperar los clusters y sus elementos

```python
# vamos a interpretar los ejementos que pertenecen a cada cluster

from scipy.cluster.hierarchy import fcluster
```

```python
Z # mi Z de 2 clusters
```

```python
max_d=20
clusters = fcluster(Z, max_d, criterion="distance")
clusters

# si vive en el cluster 1, 2 o 3 (con max_d = 20)
# podemos usar el cliterio distancia maxima
```

```python
k=2
clusters = fcluster(Z, k, criterion="maxclust")
clusters

# o por el metodo de inconsistencia (maxclust)
```

```python
fcluster(Z, 8, depth=10)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X[:,0], X[:,1], c = clusters, cmap="prism")
plt.show()
```

```python
max_d=170
clusters = fcluster(Z2, max_d, criterion="distance")
clusters

# para nuestro otro set usando max_d = 170 - 5 clusters

plt.figure(figsize=(10,8))
plt.scatter(X2[:,0], X2[:,1], c = clusters, cmap="prism")
plt.show()
```

```python
k = 4 # si usamos con maxclust y el valor que nos tiró el algoritmo
clusters = fcluster(Z2, k, criterion="maxclust")
clusters

# para nuestro otro set usando k 4 clusters

plt.figure(figsize=(10,8))
plt.scatter(X2[:,0], X2[:,1], c = clusters, cmap="prism")
plt.show()
```
